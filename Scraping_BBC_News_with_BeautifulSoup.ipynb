### Importing Libraries

```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

### Collecting Data

```python
# Url for website
url = "http://feeds.bbci.co.uk/news/rss.xml"
```

```python
# Importing the page code
reponse = requests.get(url)
soup = BeautifulSoup(reponse.text, "html.parser")
soup
```

```python
items = soup.findAll('item')
```

```python
items # list
item = items[0]
item
```

```python
item.title.text
```

```python
item.description.text
```

```python
news_items =[]
for i in items :
    news_i ={}
    news_i['title']=i.title.text
    news_i['description']=i.description.text
    news_i['pubdate']=i.pubdate.text
    news_items.append(news_i)
```

```python
news_items
```

### Saving Data

```python
# Creating a dataframe using list of data
df = pd.DataFrame(news_items, columns=['title','description','pubdate'])
```

```python
df.head(6)
```

```python
# generate csv file
df.to_csv('Scrapping_using_beautifulsoup.csv', index=False, encoding='utf-8')
```
